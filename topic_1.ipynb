```python
import pandas as pd
import numpy as np
import fyp_functions as fyp
import matplotlib.pyplot as plt
import itertools
```

```python
P = fyp.generate_random_preferences(9,2,2)

df = fyp.iterate_over_alphas(P, 20, disutility_type='util')

fig, ax = plt.subplots(figsize=(7, 6))
ax.plot(df['alpha'],df['util'],marker ='o')
        
ax.set_title('Welfare Tradeoff')
ax.set_ylabel('Welfare')
ax.set_xlabel('Alpha')
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
P = np.array([[0.86, 0.14],
       [0.77, 0.23],
       [0.85, 0.15],
       [0.94, 0.06],
       [0.02, 0.98],
       [0.77, 0.23],
       [0.56, 0.44],
       [0.67, 0.33],
       [0.5 , 0.5 ]])
df = fyp.iterate_over_alphas(P, 11, disutility_type='util')


first_values = P[:, 0]

# We'll plot all points at x=0, with y-values from first_values.
x_coords = np.zeros_like(first_values)

plt.figure(figsize=(3, 6))  # Make the figure tall to emphasize the vertical line
plt.scatter(x_coords, first_values, color='blue')

# Add labels to each point
for i, y in enumerate(first_values):
    plt.text(0.02, y, f"{y:.2f}", fontsize=10, verticalalignment='center')

# Draw a line from (0, 0) to (0, 1) for reference
plt.axvline(x=0, color='gray', linestyle='--', linewidth=1)

# Set the y-limits from 0 to 1
plt.ylim(0, 1)

plt.xticks([])

val_alpha_0 = df.loc[df['alpha'] == 0, 'allocation'].values[0][0]
val_alpha_1 = df.loc[df['alpha'] == 1, 'allocation'].values[0][0]


# Horizontal lines for alpha=0 and alpha=1
plt.axhline(y=val_alpha_0, color='red', linestyle='--', label=f'IMM: {val_alpha_0:.3f}')
plt.axhline(y=val_alpha_1, color='green', linestyle='--', label=f'WMPMN: {val_alpha_1:.3f}')

plt.title("When vote is between two allocation")
plt.legend(loc='best')
plt.show()
```
